```python
%matplotlib inline
import numpy as np
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def g(p):
    x_1, x_2 = p
    return 1/2*(2*x_1**2 + x_2**2) - 2*(2*x_1 + x_2)
```

```python
x = np.linspace(0, 2, .01)
y = np.linspace(0, 2, .01)
```

```python
#Plot set
fig = plt.figure(figsize=(15, 10))
ax = fig.gca(projection='3d')

#Plot
x_mesh = np.arange(0, 2, .1)
y_mesh = np.arange(0, 2, .1)
X, Y = np.meshgrid(x_mesh, y_mesh)
Z = (2*X**2 + Y**2) - 2*(2*X + Y)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.spectral, linewidth=0, antialiased=True)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
